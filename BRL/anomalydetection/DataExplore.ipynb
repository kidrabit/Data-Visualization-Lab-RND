```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data_list=[]
file_names=['2006-05-13.csv','2006-05-14.csv','2006-05-15.csv','2006-05-16.csv','2006-05-17.csv','2006-05-18.csv',
            '2006-05-19.csv','2007-12-14.csv','2009-05-05.csv','2011-11-01.csv']
for f in file_names:
    data=pd.read_csv('data/raw_data/'+f)
    if f in ['2007-12-14.csv','2009-05-05.csv','2011-11-01.csv']:
            q25,q75=np.quantile(data['SJS13'],[0.25,0.75])
            data['SJS13'][data['SJS13']<q25-1.5*(q75-q25)]=np.nan#3000
            data['SJS13'][data['SJS13']>q75+1.5*(q75-q25)]=np.nan#2700
    data=data.interpolate(method='linear')
    data=data.iloc[90:-90]
    #print(np.sum(np.isnan(data)))
    data_list.append(data)
    
for i in range(len(file_names)):
    fig=plt.figure(figsize=(20,5))
    plt.plot(data_list[i],'.')
    plt.grid()
    plt.legend(data_list[-1].columns)
    #plt.savefig('images/'+str(i)+'.png',bbox_inces='tight', pad_inches=0)
    plt.show()
    
fig=plt.figure(figsize=(20,5))
plt.plot(data_list[-1],'.')#,alpha=0.1)
plt.grid()
plt.legend(data_list[-1].columns)
#plt.savefig('images/11.png',bbox_inces='tight', pad_inches=0)
plt.show()
```
