메모리 체크
===========
memory_profiler 패키지로 메모리 체크

사용법
- current_memory_check() => 현재 메모리 사용량을 string 형태로 반환
- current_notebook_name() => 현재 사용중인 주피터 노트북 이름을 반환
- get_memory_usage() => |출력

```python
def current_memory_check():
    mem_usage = memory_usage(-1, interval = 1, timeout = 1)[0]
    usage = f"{mem_usage: .3f} MB"
    return usage

def current_notebook_name():
    notebook_name = ipyparams.notebook_name
    name = notebook_name + ":"
    return name

def get_memory_usage():
    print(current_notebook_name() + "" + current_memory_check())
```

일 생산 계획 수립
======================
일 생산계획 수립 데이터를 포장장별 + 날짜별로 나누는 함수 작성

```python
def getCsvFile(day):
    path = "Data\\" + dic + "\\" + file_name + "_" + day + ".csv"
    df = pd.read_csv(path)
    
    get_memory_usage()
    
    return df

def make_new_dataframe(origin_df):
    column = ['REG_YMD', 'CUST_CD', 'CUST_NM', 'ORD_NO', 'CAR_TP_CD', 'CAR_TP_NM',
          'PART_NO', 'PART_NM', 'VEND_CD', 'VEND_NM', 'DELAY_DAY', 
          'BOX_CAPA_QTY', 'CASE_CAPA_QTY', 'CAPA_QTY', 'M_BOX_CD',
          'PACK_PLAN_QTY', 'PACK_PLAN_CASE', 'PACK_SP_CD']

    new_column = ['등록일자', '고객사코드', '고객사명', '오더번호', '차종', 
                  '차종명', '부품번호', '부품명', '업체코드', '업체명', '지연일수',
                  '중박스적입수', '대박스적입수', '케이스적입수', '중박스코드',
                  '계획수량', '계획케이스', '포장장코드']
    
    df = pd.DataFrame(columns = new_column)
    
    for i, col in enumerate(column):
        df[new_column[i]] = origin_df[col]
        
    return df
```

```python
def num_of_kinds(column):
    array = []
    
    for i in range(len(plan_df)):
        array.append(plan_df.iloc[i][column])
        
    data_set = set(array)
    data_list = list(data_set)
    data_list.sort()
    
    return data_list

def get_new_dataframe():
    column = ['등록일자', '고객사코드', '고객사명', '오더번호', '차종', '차종명', 
         '부품번호', '부품명', '업체코드', '업체명', '지연일수',
         '중박스적입수', '대박스적입수', '케이스적입수', '중박스코드',
         '계획수량', '계획케이스', '포장장코드']
    
    df = pd.DataFrame(columns = column)
    return df

def make_df_list(df):
    pack = num_of_kinds('포장장코드')
    print(pack)
    date = num_of_kinds('등록일자')
    print(date)
    
    df_list = []
    for i in range(len(pack)):
        df_list.append([])
        for j in range(len(date)):
            df_list[i].append(get_new_dataframe())
            
    for i in range(len(df)):
        data = df.iloc[i]
        row = pack.index(data['포장장코드'])
        col = date.index(data['등록일자'])
        
        df_list[row][col] = df_list[row][col].append(data, ignore_index = True)
        
        if i%1000 == 0:
            print(i + "회")
        
    return df_list
```

```python
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print('Error: Creating directory. ' + directory)
        
def make_afterprocessing_data(df_list, date):
    base_dir = 'PreProcessing\Data\\' + date + '\\'
    
    for i in range(len(pack)):
        for j in range(len(date)):
            if len(df_list[i][j]) == 0:
                continue

            if j == 0:
                directory = df_list[i][j]['포장장코드'][0]
                createFolder(base_dir + directory)
            day = str(df_list[i][j]['등록일자'][0])

            path = base_dir + directory + '\\' + directory + '_' + day + '.csv'
            df_list[i][j].to_csv(path, encoding='utf-8-sig')
```

```python
df = getCsvFile('day')
plan_df = make_new_dataframe(df)
df_list = make_df_list(plan_df)
make_afterprocessing_data(df_list, 'date')
```
