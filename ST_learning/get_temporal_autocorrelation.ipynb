```python
plt.figure(figsize=(20,5))
for s in range(array.shape[1]):
    auto_corr=[]
    for i in range(1,80):
        ac=pd.Series(array[:,s,0]).autocorr(lag=i)
        auto_corr.append(ac)
    
    plt.plot(auto_corr)
    plt.xticks(np.arange(0,80,12))
    plt.grid()
    #print(i,ac)
```
