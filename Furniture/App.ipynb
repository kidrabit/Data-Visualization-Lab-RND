```python
import numpy as np
from tkinter import *
from tkinter import colorchooser
```

```python
class GridEnv():
    def __init__(self, title='Application'):
        self.window = Tk()
        self.window.geometry('800x800+500+100')
        self.window.title(title)
        self.window.resizable(width=False, height=False) #Ignore Resize
        
        #Grid Canvas
        self.cnt_grid_x = 10
        self.cnt_grid_y = 10
        self.grid_x = 400
        self.grid_y = 400
        self.recX = self.grid_x / self.cnt_grid_x
        self.recY = self.grid_y / self.cnt_grid_y
        
        self.max_grid = 30
        self.min_grid = 2
        
        self.canvas = None
        
        self.canvas2 = Canvas(self.window, width=250, height = 600, bg='cyan')
        self.canvas2.place(x=500, y=50)
        
        self.furniture_list = {}
        
        self.create_resize_grid_layout()
        self.init_grid()
        self.init_create_module()
        
    def runApp(self):
        self.window.mainloop()
        
    def create_resize_grid_layout(self):
        label = Label(self.window, text = 'RoomSize')
        label.config(font=("Courier", 14))
        label.place(x=100, y=50)
        
        label_multi = Label(self.window, text='X')
        label_multi.config(fon=("Courier", 16))
        label_multi.place(x=150, y=100)
        
        self.textbox = Text(self.window, height=1, width=7)
        self.textbox.insert(1.0, '10')
        self.textbox.configure(font=("Courier", 16))
        self.textbox.place(x=50, y=100)
        
        self.textbox2 = Text(self.window, height=1, width=7)
        self.textbox2.insert(1.0, '10')
        self.textbox2.configure(font=("Courier", 16))
        self.textbox2.place(x=170, y=100)
        
        self.submit_bt = Button(self.window, text='Submit', command = self.submit_click)
        self.submit_bt.place(x=280, y=100)
        
    def submit_click(self):
        t1 = self.textbox.get('1.0', 'end').replace('\n', '')
        t2 = self.textbox2.get('1.0', 'end').replace('\n', '')
        t1 = int(t1)
        t2 = int(t2)
        
        if t1 > self.max_grid:
            t1 = self.max_grid
            self.textbox.delete("1.0","end")
            self.textbox.insert(1.0, str(self.max_grid))
        elif t1 < self.min_grid:
            t1 = self.min_grid
            self.textbox.delete("1.0","end")
            self.textbox.insert(1.0, str(self.min_grid))
        
        if t2 > self.max_grid:
            t2 = self.max_grid
            self.textbox2.delete("1.0","end")
            self.textbox2.insert(1.0, str(self.max_grid))
        elif t2 < self.min_grid:
            t2 = self.min_grid
            self.textbox2.delete("1.0","end")
            self.textbox2.insert(1.0, str(self.min_grid))
        
        self.cnt_grid_x = t1
        self.cnt_grid_y = t2
        self.recX = self.grid_x / self.cnt_grid_x
        self.recY = self.grid_y / self.cnt_grid_y
        
        self.init_grid()
        
    def on_enter(self,event):
        item = event.widget.find_withtag('current')
        event.widget.itemconfig(item, fill='red')
        
    def on_leave(self, event):
        item = event.widget.find_withtag('current')
        event.widget.itemconfig(item, fill='white')
        
    def on_click(self, event):
        item = self.canvas.find_withtag(10)
        self.canvas.itemconfig(item, fill='black')
        
        
    def choose_color(self):
        self.color_code = colorchooser.askcolor(title='Choose color')
        self.canvas2.itemconfig(self.canvas2.find_withtag(self.color_rect), fill=self.color_code[1])
        
    def create_furniture(self):
        key = 'furniture ' + str(len(self.furniture_list))
        self.furniture_list[key] = {}
        
        name = self.name_txt_box.get('1.0', 'end').replace('\n', '')
        color = self.color_code[1]
        size_x = self.size_txt1.get('1.0', 'end').replace('\n','')
        size_y = self.size_txt2.get('1.0', 'end').replace('\n','')
        size_z = self.size_txt3.get('1.0', 'end').replace('\n','')
        size = size_x + 'x' + size_y + 'x' + size_z
        
        ismain = self.ismain.get()
        
        self.furniture_list[key]['Name'] = name
        self.furniture_list[key]['Color'] = color
        self.furniture_list[key]['Size'] = size
        self.furniture_list[key]['IsMain'] = ismain
        
        sx = int(size_x)
        sy = int(size_y)
        
        rect = Canvas(self.window, bg=color, width=(sx * self.recX)-1, height=(sy * self.recY)-1, highlightthickness=0)
        mouse_x = self.window.winfo_pointerx() - self.window.winfo_rootx()
        mouse_y = self.window.winfo_pointery() - self.window.winfo_rooty()
        rect.place(x=mouse_x, y=mouse_y)
        rect.bind('<B1-Motion>', self.moved)
        
    def moved(self, event):
        mouse_x = self.window.winfo_pointerx() - self.window.winfo_rootx()
        mouse_y = self.window.winfo_pointery() - self.window.winfo_rooty()
        event.widget.place(x=mouse_x, y=mouse_y)
    
    def init_create_module(self):
        #self.window -> self.canvas2, 상대좌표 싹다바꿔
        self.create_btn = Button(self.window, text = 'Create', width = 30, height = 2, bg='cyan', command = self.create_furniture)
        self.create_btn.place(x=520, y=50)
        
        self.name_label = Label(self.window, text='Name', bg='cyan')
        self.name_label.configure(font=("Courier", 16))
        self.name_label.place(x=510, y=130)
        self.name_txt_box = Text(self.window, height=1, width=10)
        self.name_txt_box.configure(font=("Courier", 16))
        self.name_txt_box.place(x=600, y=130)
        
        self.color_label = Label(self.window, text = 'Color', bg='cyan')
        self.color_label.configure(font=("Courier", 16))
        self.color_label.place(x=510, y=200)
        self.color_button = Button(self.window, text='Select', command = self.choose_color, height = 1, width = 10)
        self.color_code = ((255.99609375, 255.99609375, 255.99609375), '#ffffff')
        self.color_rect = self.canvas2.create_rectangle(200, 150, 230, 180, fill=self.color_code[1])
        self.color_button.place(x=600, y=200)
        
        self.size_label = Label(self.window, text='Size', bg='cyan')
        self.size_label.configure(font=("Courier", 16))
        self.size_label.place(x=510, y=300)
        self.size_txt1 = Text(self.window, height = 1, width = 3)
        self.size_txt1.configure(font=("Courier", 13))
        self.size_txt1.place(x=600, y=300)
        self.size_txt2 = Text(self.window, height = 1, width = 3)
        self.size_txt2.configure(font=("Courier", 13))
        self.size_txt2.place(x=650, y=300)
        self.size_txt3 = Text(self.window, height = 1, width = 3)
        self.size_txt3.configure(font=("Courier", 13))
        self.size_txt3.place(x=700, y=300)
        
        self.size_txt1.insert(1.0, '1')
        self.size_txt2.insert(1.0, '1')
        self.size_txt3.insert(1.0, '1')
        
        self.size_mul_label1 = Label(self.window, text='X', bg='cyan')
        self.size_mul_label1.configure(font=("Courier" ,16))
        self.size_mul_label1.place(x=630, y=300)
        self.size_mul_label2 = Label(self.window, text='X', bg='cyan')
        self.size_mul_label2.configure(font=("Courier" ,16))
        self.size_mul_label2.place(x=680, y=300)
        
        self.main_label = Label(self.window, text='Main\nFurniture', bg='cyan')
        self.main_label.configure(font=("Courier",16))
        self.main_label.place(x=510, y=350)
        
        self.ismain = IntVar()
        self.chk_btn = Checkbutton(self.window, variable=self.ismain, bg='cyan', width=3, height=3, padx=0.2, pady=0.2)
        self.chk_btn.place(x=650, y=350)
    
    def init_grid(self):
        if self.canvas != None:
            self.canvas.destroy()
        
        self.canvas = Canvas(self.window, width=self.grid_x+50, height=self.grid_y+50)
        self.canvas.place(x=20, y=200)
        
        for i in range(self.cnt_grid_x):
            for j in range(self.cnt_grid_y):
                rect_id = self.canvas.create_rectangle(i*self.recX+5, j*self.recY+5, 
                                                       (i+1)*self.recX+5, (j+1)*self.recY+5, fill = 'white', tags = 'grid', width=1)
                self.canvas.tag_bind(rect_id, '<Enter>', self.on_enter)
                self.canvas.tag_bind(rect_id, '<Leave>', self.on_leave)
                self.canvas.tag_bind(rect_id, '<Button-1>', self.on_click)
```

```python
grid = GridEnv()
grid.runApp()
```

```python
1. Create 누르면 마우스 위치에 생성 + 마우스 위치 따라다님 + 50% 투명도
2. Grid에 올라가면 Hover, 나가면 Leave
3. 클릭하면 누르는 색 바뀐 후 해당 위치에 고정 + 100% 투명도
4. 우클릭하면 삭제

클릭시 -> item = canvas.find_withtag(id) => thisrect.setposition(item.position)
방 변화 클릭시 -> grid창 맨 아래로 보내기
중복된 위치 체크 모듈
```

```python
grid.window.destroy()
```
