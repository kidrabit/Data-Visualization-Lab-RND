```python
def load_pickle(filename):
    path = 'data/' + filename + '.p'
    with open(path, 'rb') as file:
        x = pickle.load(file)
    return x
```

categories_dict.p
- room image : room type

categories_images_dict.p
- obj image : obj type

img_to_desc.p
- obj image : {desc : 설명, name: type}

item_to_room.p
- object image가 어느 room image에 있는지

products_dict.p
- object의 이름, type, 색, 크기, img이름, 설명

room_to_items.p
- room image : 해당 이미지에 무엇이 있는지

```python
cat = load_pickle('categories_dict')
cat_img = load_pickle('categories_images_dict')
img = load_pickle('img_to_desc')
item = load_pickle('item_to_room')
product = load_pickle('products_dict')
room = load_pickle('room_to_items')
```
