```python
import cv2
```

```python
method = cv2.TM_SQDIFF_NORMED
```

```python
small_img = cv2.imread('test_data/test5.jpg')
large_img = cv2.imread('test_data/test2.jpg')
```

```python
cv2.TM_SQDIFF_NORMED
```

```python
import matplotlib.pyplot as plt
plt.imshow(small_img)
```

```python
plt.imshow(large_img)
```

```python
result = cv2.matchTemplate(small_img, large_img, method)
```

```python
result
```

```python
mn, mx, mnLoc, mxLoc = cv2.minMaxLoc(result)
```

```python
mn, mnLoc
```

```python
MPx, MPy = mnLoc
trows, tcols = small_img.shape[:2]
cv2.rectangle(large_img, (MPx, MPy), (MPx+tcols, MPy+trows),(0,0,255),2)
cv2.imshow('output', large_img)
cv2.waitKey(0)
```
