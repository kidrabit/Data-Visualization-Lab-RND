```python
import os
import json
import numpy as np
import pandas as pd
```

```python
types = {'Bed':1, 'Table':2, 'Chair':3, 'Shelf':4, 
         'Closet':5, 'Cabinet':6, 'Sofa':7, 
         'Bookshelf':8,'Trash':9, 'Pot':10, 'TV':11, 
         'Computer':12, 'Mirror':13, 'Hanger':14, 'Lamp':15}
```

```python
path = 'data/make/'

datalist = []
for file in os.listdir(path):
    with open(path + file) as json_file:
        json_data = json.load(json_file)
    datalist.append(eval(json_data))
```

```python
datas = []
cnts = []
for i in range(10):
    datas.append([])
    cnts.append([0 for i in range(16)])
    for j in range(20):
        datas[i].append([])
        for k in range(20):
            t = datalist[i]['data'][j][k]['type']
            if t == '':
                datas[i][j].append(0)
            else:
                datas[i][j].append(types[t])
                cnts[i][types[t]-1] += 1
```

```python
path = 'data/target/'

targetlist = []
for file in os.listdir(path):
    with open(path + file) as json_file:
        json_data = json.load(json_file)
    targetlist.append(eval(json_data))
```

```python
targets = []
cnts_t = []
for i in range(10):
    targets.append([])
    cnts_t.append([0 for i in range(16)])
    for j in range(20):
        targets[i].append([])
        for k in range(20):
            t = targetlist[i]['data'][j][k]['type']
            if t == '':
                targets[i][j].append(0)
            else:
                targets[i][j].append(types[t])
                cnts_t[i][types[t]-1] += 1
```

```python
def print_ds(d):
    for i in range(20):
        for j in range(20):
            print('{:2d}'.format(d[i][j]), end = ' ')
        print('')
```

```python
for i in range(10):
    print('----------------', i,  '----------------')
    for j in range(16):
        if cnts[i][j] != 0:
            print(j + 1, end = ', ')
    print('')
    print_ds(datas[i])
```

```python
for i in range(10):
    print('----------------', i,  '----------------')
    for j in range(16):
        if cnts_t[i][j] != 0:
            print(j + 1, end = ', ')
    print('')
    print_ds(targets[i])
```

```python
datas = np.array(datas)
targets = np.array(targets)
datas.shape, targets.shape
```

```python
np.save('data/data', datas)
np.save('data/target', targets)
```
