```python
ccc_causality = pd.DataFrame(columns = ['Cause', 'Effect', 'score'])
from ETC import CCC
edge = cyto_full.columns
for i in range(len(edge)):
    edge1 = edge[i]
    df1 = cyto_full[edge1].values
    for j in range(len(edge)):
        edge2 = edge[j]
        df2 = cyto_full[edge2].values
        ccc_est = CCC.compute( df1, df2, LEN_past=100, ADD_meas=100, STEP_size=100, n_partitions=2)
        if abs(ccc_est) > 0.01:
          df_ccc = pd.Series([edge1, edge2, ccc_est], index = ['Cause', 'Effect', 'score'])
          ccc_causality = ccc_causality.append(df_ccc, ignore_index = True)
```

```python
ccc_causality_lorenz = pd.DataFrame(columns = ['Cause', 'Effect', 'score'])
from ETC import CCC
edge = lorenz_96.columns
for i in range(len(edge)):
    edge1 = edge[i]
    df1 = lorenz_96[edge1].values
    for j in range(len(edge)):
        edge2 = edge[j]
        df2 = lorenz_96[edge2].values
        ccc_est = CCC.compute( df1, df2, LEN_past=10, ADD_meas=2, STEP_size=2, n_partitions=2)
        if abs(ccc_est) > 0.01:
          df_ccc = pd.Series([edge1, edge2, ccc_est], index = ['Cause', 'Effect', 'score'])
          ccc_causality_lorenz = ccc_causality_lorenz.append(df_ccc, ignore_index = True)
```
