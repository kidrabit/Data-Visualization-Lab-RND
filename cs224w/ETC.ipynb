```python
import pandas as pd
data = pd.read_csv('/content/gang_csv.csv')
data =data.drop('1210030000',axis = 1)
```

```python
df_morning = data.loc[1560:1571]
```

```python
edge = df_morning.columns
```

```python
edge
```

```python
import numpy as np
```

```python
CCC_num = np.zeros((57, 57))
```

```python
CCC_num.shape
```

```python
from ETC import CCC
for i in range(57):
    edge1 = edge[i]
    df1 = df_morning[edge1].values
    for j in range(57):
        edge2 = edge[j]
        df2 = df_morning[edge2].values
        ccc_est = CCC.compute( df1, df2, LEN_past=2, ADD_meas=2, STEP_size=2, n_partitions=False)
        CCC_num[i][j] = ccc_est
```

```python
sum(CCC_num)
```

```python
np.save('CCC_num', CCC_num)
```
