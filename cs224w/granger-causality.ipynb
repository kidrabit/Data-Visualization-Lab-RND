# granger causality 개념

 ### correlation의 문제점
  - 상관관계를 보고 논리적 순서를 따지는 것은 위험이 크다.
  - 예를 들어 아래의 그림처럼 상어의 공격과 아이스크림 판매량의 상관성을 확인해 보았을 때 둘은 밀접한 연관이 있는것으로 보인다.
  
  - 하지만 이것은 여름에 사람들이 해변가에 많이 놀러가고 여름에 아이스크림을 많이 먹기 때문이다.(이 둘은 연관성이 없다.)
  - 이러한 문제점 때문에 causality를 살펴보아야 된다.

 ### Granger cauality
  - idea : 원인 결과가 시간 순서에 따라 발생했다면, X와 Y의 시간의 선후가 서로에게 영향을 주었는지 살펴보는 것이다.
  - ex) 
      - $X_t, Y_t, W_t$의 세 가지 항이 있다고 가정($W_t $:  a vector of possible explanatory variables)
      - 먼저 $X_t$와 $W_t$의 과거 항을 사용하여 $X_{t+1}$을 예측
      - 다음 $X_t$, $Y_t$ 및 $W_t$의 과거 항을 사용하여 $X_{t+1}$을 예측
      - standard cost functions를 통해 두 번째 예측이 더욱 유의미 하면 Y의 과거 항들이 $X_t,W_t$에 정보보다 $X_{t+1}$을 예측하는데 유의미 하다는 뜻이다.
      - 만약 $Y_t$가 $X_{t+1}$ 보다 먼저 발생했다면(시간) $Y_t$는 "Granger cause"이 된다.(다른 그룹에 발견되지 않는 유용한 정보를 기자고 있는)
      - $W_t $가 크고 and the more carefully its contents are selected 수록 $Y_t$는 더욱 엄격히 평가된다.
      

 ## Granger cauality test
  - 일반적으로 linear regression models을 사용한다.(t검정과 F검정 사용)
  - 변수 X1과 X2을 예로들어 보면 아래와 같은 식을 계산한다.
  
  - p : 관측치의 최대의 수, A: model coefficients, E1,E2 : 시계열의 잔차(예측오차)
  - 만약 위의 식에서 E1(or E2)의 variance가 감소하면 X2(0r X1)가 X1(or X2)granger cauality가 된다고 볼 수있다.
  - (즉 E1's variance reduced는 X2 -> X1으로 granger cauality라고 말할수 있음)
  -  In other words, X2 G-causes X1 if the coefficients in A12 are jointly significantly different from zero.
  - 이는 X1과 X2 공분산 측위성의 가정을 $A_12$ = 0 으로 두고 F-검정을 진행할수 있다. 
  - 두개의 변수 말고도 변수를 추가하여 다변량 변수들도 추론할수있다.

## Limitations and extensions
 - geanger cauality의 공식은 선형특징에 대한 정보만 제공한다.
 - 이를 비선형에 대입하기 위해 radial basis function을 사용
 
 #### Radial basis function(거리에만 의존하는 함수?)...((1)의 해결)
 - 어떤 함수에 대한 근사(Approximation)모델을 Radial Function의 선형조합으로 표현할 수 있다
  - single univariate function에 기초한 terms들의 linear combination으로 multivariate function에 근사한 평균이다.
  
  - 이는 둘 이상의 차원으로 사용할수 있게 radialised 되어있다.
  - finite number of points로 알려진 data에서 사용
  - neural networks에서  radially symmetric functions which are shifted by points in multidimensional Euclidean space and then linearly combined, they form data-dependent approximation spaces. 
  
  - Definition of the method
      - 유클리드 공간 $R^n$에서 점들($x_1,...,x_n$)은 양수이고 서로 다르다고 가정한다.
  
      - scalar parameters는 점 $x_j$를 제외한 모든 지점에서 s를 f에 잘 근사되게 선택한다.
      - 그러므로 일반적으로 잘 알려져 있지 않은 function f는 linear expression식에 의해 근사된다.
  
  - 나의 이해 : (비선형을 선형으로 근사 시킨후 사용하기 위함인거 같다.)
