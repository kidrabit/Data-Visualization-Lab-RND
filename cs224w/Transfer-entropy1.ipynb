# Transfer Entropy (TE)

 - Aussuming
     - $I_{(n+1)}$ 과 $J_{(n)}$은 independence
     - distribution : $q = p(i_{(n+1)} | i_n^k)$

 - If I and J are independent processes, then $p(i_{n+1},i_n^k,j_n^l)$ = $p(i_{n+1},i_n^k)$ for all n, k,l and hence the above quantity will be zero. Intuitively, $TE_{J→I}$ captures the flow of information from a process J to a process I.
