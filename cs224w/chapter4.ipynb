# community structure
 - network 상에서 densely하게 연결되어 있는 node_set을 말한다. -> 자동적으로 community structure을 찾는 알고리즘 필요
 - community structure가 생기는 이유
   - Granovetter's Answer(points : structure, information)
   - 우리가 유용한 정보들을 친한 친구에게 보다는 지인들에게 잘 얻는것에서 파생됨.
   - 1) link_1 : strong (친한친구)
   - 2) link_2 : week (지인)
   - link_1의 경우 강한 연결성을 띄므로 network상 short-edge를 가지고 있고, 중복된 정보들을 담고있다.
   - link_2의 경우 약한 연결성을 띄므로 network상 long-edge(or 두 network를 연결)를 가지고 있고, 새로운 정보를 담고있다.
   
### Edge overlap
 - network 안에서 nodes들의 공유 수치
 - $O_{ij} = \lvert (N(i) and N(j)) / (N(i) or N(j))\lvert$ (N(i) is the set of neighbors of node_i)
 - common neighbor를 공유할수록 값이 1에 가까움

### network communities
 - set of nodes with lots of internal connections and few external ones
 #### Modilarity Q
   - network가 community로 얼마나 잘 나누어져 있는가에 대한 수치
   - Q =  (#edges within group s) - (expected # edges within group s)  -> expected # edges within group s를 찾는 모델이 필요(Configuration Model)
   - 아래와 같이 # edges within group s를 구하였다면, Q(G,S) = Q/2m으로 정의 할수 있다.-> -1~1사이의 값 
 #### Configuration Model
   - real graph G가 주어졌을 때 G와같은 node number와 edge number를 가지는 random graph G'를 그린다.
   - random graph에서 node_i와 node_j의 degree가 k_i, k_j일때 # edges within group s : k_i * (k_j/2m) (m : all edge number)
   
### Louvain Algorithm
 - communities in network를 빠르게 찾을 수있게 하는 알고리즘
 - Greedy algorithm의 일종
 - Louvain Algorithm은 modilarity를 최대화하는 것이 목적이다.
   - 과정
     - 가장 처음은 각각의 노드가 single community라고 정의한다.
     - 노드의 community를 계속 바꿔나가면서 modularity의 증가량을 살펴본다.
     - 더이상 modularity가 개선되지 않는다면 종료한다.
     - 위에서 찾은 community를 하나의 single super-node를 만든다.
     - 다시 위의 과정을 반복한다.
     - ΔQ = ΔQ(i→C)+ΔQ(D→i) :ΔQ(i→C):노드 i를 'C' community 추가시켰을 때 Q의 증가량,ΔQ(D→i):'D' community 에서 노드i를 제거시켰을   Q의 증가량

### Detecting Overlapping Communities
 - 모든 network가 중복되어 있는 지역이 없는 것은 아니다.
 - 즉, overlap되어 있는 부분이 존재 할것이다.
 - 따라서 overlap되어 있는 부분을 찾아야된다.
    - 과정
       - define a generative model for graphs that is based on node community affiliations (AGM) -> 아직 배우지 않음
       - Given graphs G , make the assumption that G was generated by AGM
       - find the best AGM that could have generated G(best parameter 찾기?)
       - 위의 모델 파라미터는 node가 community에 얼마나 속해 있을지에 대한 것이라고 한다.
 - coummunity Affiliation -> model -> network
    - generative model : coummunity Affiliation 를 가지고 network를 발생 (how?)
    - model parameters : nodes V, coummunities C , Memberships M(edge)
      - 각 coummunities c는 single probability $p_c$를 가지고 있다.(node와 community c가 연결될 확률)
      - $p(u,v) = 1- (*(1 - p_c))$ : node_u,node_v가 같은 coummunity의 있을 확률
 - 지금까지 coummunity Affiliation에서 network를 만드는 과정을 했지만, 실제로는 주어진 graphs에서 coummunity Affiliation을 만들어 내는 model을 찾아야된다.
 - 주어진 graph G에서 model F의 paramerte 찾는 방법
   - 과정(maximum likelihodd estimation 사용)
      - $argmax_FP(G|F)$을 풀어야됨
      - Given a network G(V,E) we maximize $l(F)$
          - $l(F) = \sum_{(u,v) \in E} log(1-exp(F_uF_v^T)) - \sum_{(u,v) \notin E} F_uF_v^T$ ($F_u$:노드 u가 각각의 커뮤니티에 속할 확률을 가진 vector)
          - 즉, 위 식의 likelhood를 최대로 만드는 F를 찾아야된다.
      - optimization:
          - start with random F
          - update $F_{uC}$ for node u while fixing the memberships of all other nodes
          - updating takes linear time in the degree of u
